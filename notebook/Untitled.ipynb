```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
visited_urls = set()
```

```python
def get_links(url):
    retries = 0
    while retries < max_retries:
        try:
            response = requests.get(url)
            soup = BeautifulSoup(response.content, 'html.parser')
            links = soup.find_all('a')
            list_url = []
            for link in links:
                href = link.get('href')
                if href is not None and (href.startswith("/wiki/") or "id.wikipedia.org/wiki/" in href):
                    if not(href in visited_urls):
                        list_url.append(href)
                        visited_urls.add(href)
            return list_url
        except requests.exceptions.RequestException:
            retries += 1
            print("Error: retrying in {} seconds...".format(retry_delay))
            time.sleep(retry_delay)
    print("Error: maximum number of retries reached.")
    return []
```

```python
start_url = "https://id.wikipedia.org/wiki/Joko_Widodo"
max_retries = 3
retry_delay = 5
```

```python
urls_to_crawl = [start_url]
```

```python
while urls_to_crawl:
    url = urls_to_crawl.pop(0)
    if(url.startswith("/wiki/")):
        links = get_links("https://id.wikipedia.org" + url)
    else:
        links = get_links(url)
    urls_to_crawl += links
    print("Visited:", url, end=" ")
    print(rf"({len(visited_urls)})")
```

```python
import re

def has_extension_or_pattern(string):
    if re.search(r'\.[a-zA-Z0-9]+$', string):
        return True
    elif re.search(r'/wiki/(\w+):', string):
        return True
    else:
        return False
```

```python
has_extension_or_pattern('/wiki/Berkas:Wakil_Gubernur_Sulawesi_Tengah,_Rusli_Dg._Palabbi.png')
```

```python
from tqdm import tqdm
cleaned_dataset = []
for i, url in tqdm(enumerate(list(visited_urls))):
    if not(has_extension_or_pattern(url)):
        cleaned_dataset.append(url)
```

```python
def get_table(url, name):
    
    max_retries = 5

    for i in range(max_retries):
        try:
            response = requests.get(url)
            html_content = response.content
            soup = BeautifulSoup(html_content, 'html.parser')
            table = soup.find('table', class_='wikitable')
            if table:
                df = pd.read_html(str(table))[0]
                header = df.iloc[:2]
                header.columns = header.columns.map(''.join)
                header = header.loc[:, ~header.columns.duplicated()]
                df = df.iloc[2:]
                df.columns = header.columns
                df.to_csv(rf'csv/{name}.csv', index=False)
                df.to_html(rf'csv/{name}.html', index=False)
                df.to_csv(rf'csv/{name}.tsv', sep='\t', index=False)
                df.to_markdown(rf'csv/{name}.table', index=False)
                data_csv.append((url, name))
                print(len(data_csv))
                return True
            else:
                return False
        except requests.exceptions.ConnectionError:
            print(f"Connection error occurred. Retrying in 5 seconds ({i+1}/{max_retries})...")
            time.sleep(5)
    return
```
